```python
import pandas as pd


df = pd.read_csv("gold_prices.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head()
```

```python
df.ffill(inplace=True)
df.bfill(inplace=True)
df['Moving_Avg'] = df['Close'].rolling(window=7).mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['Open', 'High', 'Low', 'Volume']]
y = df['Close']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


model = LinearRegression()
model.fit(X_train, y_train)


predictions = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

# Plot the actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(y_test.index, y_test, label='Actual Prices', color='blue')
plt.plot(y_test.index, predictions, label='Predicted Prices', color='red', linestyle='--')
plt.title('Gold Prices: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, predictions)
print(f'Mean Absolute Error: {mae}')
```

```python
import seaborn as sns

# Visualize the Actual vs Predicted
plt.figure(figsize=(12, 6))
sns.lineplot(data=df.loc[y_test.index], x=y_test.index, y='Close', label='Actual Prices', color='blue')
sns.lineplot(x=y_test.index, y=predictions, label='Predicted Prices', color='red', linestyle='--')
plt.title('Gold Price Prediction: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
